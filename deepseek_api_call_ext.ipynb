### Functions for API Call, Run Experiments, Save Results

```python

# Initialize the client
client = OpenAI(api_key=API_KEY, base_url=BASE_URL)

def make_api_call(system_prompt, user_question, stream=False):
    """
    Make an API call to the LLM with the provided system prompt and user question.
    
    Args:
        system_prompt (str): The system prompt to use
        user_question (str): The user question to evaluate
        stream (bool): Whether to stream the response
        
    Returns:
        dict: Response data including the model's response and metadata
    """
    try:
        response = client.chat.completions.create(
            model=MODEL,
            temperature=TEMPERATURE,
            messages=[
                {"role": "system", "content": system_prompt},
                {"role": "user", "content": user_question},
            ],
            stream=stream
        )
        
        result = {
            "system_prompt": system_prompt,
            "user_question": user_question,
            "model_response": response.choices[0].message.content,
            "timestamp": datetime.now().isoformat(),
            "model": MODEL,
            "finish_reason": response.choices[0].finish_reason,
            "usage": {
                "prompt_tokens": response.usage.prompt_tokens,
                "completion_tokens": response.usage.completion_tokens,
                "total_tokens": response.usage.total_tokens
            }
        }
        
        return result
    
    except Exception as e:
        print(f"Error making API call: {e}")
        return {
            "system_prompt": system_prompt,
            "user_question": user_question,
            "error": str(e),
            "timestamp": datetime.now().isoformat()
        }

def run_experiments():
    """
    Run the full set of experiments with all system prompts and evaluation questions.
    
    Returns:
        list: All experiment results
    """
    results = []
    
    total_experiments = len(system_prompts) * len(evaluation_questions)
    completed = 0
    
    for prompt in system_prompts:
        for question in evaluation_questions:
            print(f"Testing prompt: \"{prompt[:6]}...\" with question: \"{question[:6]}...\"")
            result = make_api_call(prompt, question)
            results.append(result)
            
            completed += 1
            print(f"Progress: {completed}/{total_experiments} ({(completed/total_experiments)*100:.1f}%)")
    
    return results

def save_results(results, filename=OUTPUT_FILE):
    """Save results to a JSON file"""
    with open(filename, 'w') as f:
        json.dump(results, f, indent=2)
    print(f"Results saved to {filename}")

def load_results(filename=OUTPUT_FILE):
    """Load results from a JSON file"""
    try:
        with open(filename, 'r') as f:
            return json.load(f)
    except FileNotFoundError:
        print(f"File {filename} not found")
        return []

def add_system_prompt(new_prompt):
    """Add a new system prompt to test"""
    if new_prompt not in system_prompts:
        system_prompts.append(new_prompt)
        print(f"Added new system prompt: \"{new_prompt[:6]}...\"")
    else:
        print("This prompt already exists in the test set")

def add_evaluation_question(new_question):
    """Add a new evaluation question"""
    if new_question not in evaluation_questions:
        evaluation_questions.append(new_question)
        print(f"Added new evaluation question: \"{new_question[:6]}...\"")
    else:
        print("This question already exists in the test set")

def test_single_combination(prompt_idx, question_idx):
    """Test a single prompt-question combination"""
    if 0 <= prompt_idx < len(system_prompts) and 0 <= question_idx < len(evaluation_questions):
        prompt = system_prompts[prompt_idx]
        question = evaluation_questions[question_idx]
        
        print(f"Testing single combination:")
        print(f"System prompt: {prompt}")
        print(f"User question: {question}")
        
        result = make_api_call(prompt, question)
        print("\nModel response:")
        print(result["model_response"])
        print(f"\nTokens used: {result['usage']['total_tokens']}")
        
        return result
    else:
        print("Invalid indices. Please check your prompt and question indices.")
        return None
```
